## Principal Component Analysis




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat('ex7data1.mat')
X = data['X']

plt.scatter(X[:, 0], X[:, 1],linewidths=3,c='m')
plt.axis([0.5, 6.5, 2, 8])

```

```python
def pca(X):
    
    m, n = X.shape
    U = np.zeros(n)
    S = np.zeros(n)

    Sigma = (1 / m) * (X.T.dot(X))
    U, S, V = np.linalg.svd(Sigma)
    
   
    return U, S
```

```python
def featureNormalize(X):
    
    mu = np.mean(X, axis=0)
    X_norm = X - mu

    sigma = np.std(X_norm, axis=0, ddof=1)
    X_norm /= sigma
    return X_norm, mu, sigma
```

```python
#  Before running PCA, it is important to first normalize X
X_norm, mu, sigma = featureNormalize(X)

#  Run PCA
U, S = pca(X_norm)
print(U)
print("\n")
print(S)

#  Draw the eigenvectors centered at mean of data. These lines show the
#  directions of maximum variations in the dataset.
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1],c='m',linewidths=3)

for i in range(2):
    ax.arrow(mu[0], mu[1], 1.5 * S[i]*U[0, i], 1.5 * S[i]*U[1, i],head_width=0.25)

ax.axis([0.5, 6.5, 2, 8])
ax.set_aspect('equal')
```

```python
def projectData(X, U, K):
    
    Z = np.zeros((X.shape[0], K))
    Z = np.dot(X,  U[:, :K])
    
    return Z
```

```python
#  Project the data onto K = 1 dimension
K = 1
Z = projectData(X_norm, U, K)
print('Projection of the first example: {:.6f}'.format(Z[0, 0]))
print('(this value should be about    : 1.481274)')
```

```python
def recoverData(Z, U, K):
    
    X_rec = np.zeros((Z.shape[0], U.shape[0]))

    X_rec = Z.dot(U[:, :K].T)

    return X_rec
```

```python
X_rec  = recoverData(Z, U, K)
# print(X_rec)
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(X_norm[:, 0], X_norm[:, 1], 'bo', ms=8, mec='b', mew=0.5)
ax.set_aspect('equal')
plt.axis([-3, 2.75, -3, 2.75])
ax.scatter(X_rec[:, 0], X_rec[:, 1],c='r')
```

```python
def displayData(X, example_width=None, figsize=(10, 10)):
    
    if X.ndim == 2:
        m, n = X.shape
    elif X.ndim == 1:
        n = X.size
        m = 1
        X = X[None]  # Promote to a 2 dimensional array
    else:
        raise IndexError('Input X should be 1 or 2 dimensional.')

    example_width = example_width or int(np.round(np.sqrt(n)))
    example_height = int(n / example_width)

    # Compute number of items to display
    display_rows = int(np.floor(np.sqrt(m)))
    display_cols = int(np.ceil(m / display_rows))

    fig, ax_array = plt.subplots(display_rows, display_cols, figsize=figsize)
    fig.subplots_adjust(wspace=0.025, hspace=0.025)

    ax_array = [ax_array] if m == 1 else ax_array.ravel()

    for i, ax in enumerate(ax_array):
        ax.imshow(X[i].reshape(example_height, example_width, order='F'), cmap='gray')
        ax.axis('off')
```

```python
data = loadmat('ex7faces.mat')
X = data['X']
print(X.shape)
```

```python

X_norm, mu, sigma = featureNormalize(X)
U, S = pca(X_norm)
```

```python
K = 1
Z = projectData(X_norm, U, K)

print('The projected data Z has a shape of: ', Z.shape)
```

```python
X_rec  = recoverData(Z, U, K)
print(X_rec.shape)
displayData(X_norm[:100, :], figsize=(6, 6))
plt.gcf().suptitle('Original faces')

# Display reconstructed data from only k eigenfaces
displayData(X_rec[:100, :], figsize=(6, 6))
plt.gcf().suptitle('Recovered faces')
pass
```
