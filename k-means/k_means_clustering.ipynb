```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from matplotlib import style
import matplotlib as mpl
```

```python
def closest_centroid_given_sample(X, centroids):
    
    k = centroids.shape[0]
    idx = np.zeros(X.shape[0], dtype=int)
    for i in range(idx.size):
        J = np.sqrt(np.sum(np.square(X[i] - centroids), axis = 1))
        idx[i] = np.argmin(J)
    
    return idx
```

```python
data = loadmat( 'ex7data2.mat')
X = data['X']
print(X.shape)
K = 3  
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])    # random initialisation

idx = closest_centroid_given_sample(X, initial_centroids)       # it will tell centroid of given sample belong to
print(idx.shape)
plt.style.use('seaborn')
plt.scatter(X[:,0],X[:,1],c='y')   # inialial data without clustering
plt.show()
```

```python
def update_centroid_with_mean(X, idx, K):
    
    m, n = X.shape
    centroids = np.zeros((K, n))

    for i in range(K):
        centroids[i] = np.mean(X[idx == i], axis = 0)
        
    return centroids
```

```python
centroids = update_centroid_with_mean(X, idx, K)
print(centroids)
```

```python
def kMeans(X, initial_centroids, iteration, plot_pro=False):
    
    K = initial_centroids.shape[0]
    idx = None
    if plot_pro:
        fig,axis=plt.subplots(5,2,figsize=(14,18))        # 5,2 must be changed with iteration
#     print(axis)
        fig.suptitle('Cluster with iteration')
    for i in range(iteration):
        idx = closest_centroid_given_sample(X, initial_centroids)        # getting centroid for all
#         print(i)
        if plot_pro:
            
            if i>=5:
                
                axis[i%5][1].scatter(X[idx==0,0],X[idx==0,1],c='b')
                axis[i%5][1].scatter(X[idx==1,0],X[idx==1,1],c='m')
                axis[i%5][1].scatter(X[idx==2,0],X[idx==2,1],c='y')
                axis[i%5][1].scatter(initial_centroids[:,0],initial_centroids[:,1],c='k',marker='x',linewidths=20)
                axis[i%5][1].set_title('%d Iteration' % i)
            else:
                
                axis[i%5][0].scatter(X[idx==0,0],X[idx==0,1],c='b')
                axis[i%5][0].scatter(X[idx==1,0],X[idx==1,1],c='m')
                axis[i%5][0].scatter(X[idx==2,0],X[idx==2,1],c='y')
                axis[i%5][0].scatter(initial_centroids[:,0],initial_centroids[:,1],c='k',marker='x',linewidths=20)
                axis[i%5][0].set_title('%d Iteration' % i)
                
                
            
#             axis[i].show()
             
        initial_centroids = update_centroid_with_mean(X, idx, K)             # updating position of centroid

    if plot_pro:
        
        plt.scatter(X[idx==0,0],X[idx==0,1],c='b')
        plt.scatter(X[idx==1,0],X[idx==1,1],c='m')
        plt.scatter(X[idx==2,0],X[idx==2,1],c='y')
        plt.scatter(initial_centroids[:,0],initial_centroids[:,1],c='k',marker='x',linewidths=20)
        plt.show()

    return initial_centroids, idx
```

```python
K = 3
iteration= 10

initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
centroids, idx = kMeans(X, initial_centroids,iteration, True)
print(centroids,idx)
```

###  Image compression with K-means

for intuition
https://towardsdatascience.com/image-compression-using-k-means-clustering-aa0c91bb0eeb

```python
def kMeansInitCentroids(X, K):
    
    m, n = X.shape
    
    centroids = np.zeros((K, n))
    random_ix = np.random.permutation(X.shape[0])
    centroids = X[random_ix[:K], :]
    
    return centroids
```

```python
K = 32
iteration = 10
# short intuition
# image pixel combination=256*256*256=16,777,216  but human can't see whole , algorithm take adv. of this 
# and algorithm assignment each pixel among the k centroid or can say algo. cluster similar color to single color


A = mpl.image.imread('bird_small.png')

A /= 255
print(A.shape)
X = A.reshape(-1, 3)
print(X.shape)

initial_centroids = kMeansInitCentroids(X, K)

centroids, idx = kMeans(X, initial_centroids,iteration,False)
print(centroids.shape)
print(idx.shape)
X_recovered = centroids[idx, :].reshape(A.shape)

fig, axis = plt.subplots(1, 2, figsize=(8, 4))
axis[0].imshow(A*255)
axis[0].set_title('Original')
axis[0].grid(False)

axis[1].imshow(X_recovered*255)
mpl.image.imsave('compressed.png',X_recovered)
axis[1].set_title('Compressed, with %d colors' % K)  # set title for putting varible
axis[1].grid(False)
```
